```python
from tqdm import tqdm
import pandas as pd
import numpy as np

csv_list = [
  'transactions_train',
  'articles',
  'customers',
  'sample_submission'
]

# for f_name in csv_list:
#   f_csv = f"data/{f_name}.csv"
#
#   print(f"Loading {f_csv}")
#   df = pd.read_csv(f_csv)
#   f_pqt = f"data/{f_name}.parquet"
#
#   print(f"Saving {f_pqt}")
#   df.to_parquet(f_pqt, engine="pyarrow")

from time import time

dfs = {}

for f_name in csv_list:
  start = time()
  f_pqt = f"data/{f_name}.parquet"
  print(f"Loading {f_pqt}")
  dfs[f_name] = pd.read_parquet(f_pqt, engine="pyarrow")
  end = time()
  print(f"Time = {end - start}")

articles_df = dfs["articles"]
transactions_df = dfs["transactions_train"]
customers_df = dfs["customers"]
```

```python
small_transaction_df = transactions_df[small_transaction_df_locs]
```

```python
transactions = pd.merge(small_customers_df, small_transaction_df, on='customer_id', how='inner')
transactions_articles = pd.merge(transactions, articles_df, on='article_id', how='inner')
```
